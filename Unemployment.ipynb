```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('Unemployment.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Step 4: Check for missing values
missing_values = data.isnull().sum()

# If necessary, fill or drop missing values
# Example: Drop rows with missing values
data_cleaned = data.dropna()

# Display missing values to confirm
missing_values

```

```python
# Step 5: Analyze national unemployment rate over time
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data_cleaned,)
plt.title('Unemployment Rate Over Time (National Level)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
